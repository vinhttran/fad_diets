```python
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
tweets_data_path = 'beyonce.txt'

tweets_data = []
tweets_file = open(tweets_data_path, "r")
for line in tweets_file:
    try:
        tweet = json.loads(line)
        tweets_data.append(tweet)
    except:
        continue
```
